We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
Expected Output:

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
Max_temp =np.array( [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25 ])
Min_temp =np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18 ])
```

```python
months = np.arange(1,13)
```

```python
plt.figure(figsize=(8,5),facecolor='r')
plt.plot(months,Max_temp,'o-')
plt.plot(months,Min_temp,'ro-')
plt.legend(['Max_temp','Min_temp'])
plt.title('Temperature of city in india',)
plt.xlabel('Months')
plt.ylabel('Min and Max Temperature')
```

# 1.fitting it to the periodic function

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,Max_temp, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,Min_temp, [-40, 20, 0])
```

# Plotting to fit

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, Max_temp, 'go')
plt.plot(days, yearly_temps(days, *res_max), 'm-')
plt.plot(months, Min_temp, 'co')
plt.plot(days, yearly_temps(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
